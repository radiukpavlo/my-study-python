```python
import numpy as np
import matplotlib.pyplot as plt

# Data
labels = ['Sunny (the Sun is \n at its zenith)', 'Sunny (high shade)', 'Cloudy']
UAV_1 = [1.7485, 1.7298, 1.6314]
UAV_2 = [1.7124, 1.7178, 1.6319]
UAV_3 = [1.7201, 1.7197, 1.6393]

x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots(figsize=(8,5))
plt.rcParams.update({'font.size': 10, 'font.weight': 'bold'})

# Soft colors for the bars
colors = ['tab:red', 'limegreen', 'dodgerblue']

# Create bars
rects1 = ax.bar(x - width, UAV_1, width, label='UAV_1', color=colors[0])
rects2 = ax.bar(x, UAV_2, width, label='UAV_2', color=colors[1])
rects3 = ax.bar(x + width, UAV_3, width, label='UAV_3', color=colors[2])

# Set y-axis limits
ax.set_ylim(1, 2)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Values', fontweight='bold', fontsize=12)
ax.set_title('Values by UAV and Weather Condition', fontweight='bold', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontweight='bold', fontsize=12)
ax.legend(loc='upper center', ncol=3)

# Autolabel function to display the label on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontweight='bold')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# Make y-axis labels bold
# ax.yaxis.set_tick_params(labelsize='medium')
# plt.yticks(fontweight='bold')

fig.tight_layout()

plt.show()
```
