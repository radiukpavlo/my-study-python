### Параметричні графіки
$$
\large x = \sin t/t,
\quad
\large y = \cos t/t
$$
$x = \frac{\sin t}{t},\ y = \frac{\cos t}{t}$

````python
t = np.linspace(1.5, 32, 200)
plt.axis('equal')
plt.title(r'$x = \frac{\sin t}{t},\ y = \frac{\cos t}{t}$')
x = np.sin(t)/t
y = np.cos(t)/t
plt.plot(x, y)
````

## Команда помилок

````python
help(plt.errorbar)
````

````python
x = np.linspace(0, 2, 11)
y = x**2
plt.errorbar(x, y, xerr = 0.1, yerr = 0.4)
````

````python
N = 11
x = np.linspace(0, 2, N)
y = x**2
x_err = x*0.1
low_y_err   = np.linspace(2, 0, N)
upper_y_err = 3*np.linspace(2, 0, N)
y_err = [low_y_err, upper_y_err]
plt.errorbar(x, y, xerr=x_err, yerr=y_err, marker='o', linestyle=' ', color='magenta', mec='magenta')
````

## Тип графіків "Scatter plot"

````python
plt.scatter([1, 2, 3, 4], [5, 6, 2, 3], color=['r', 'g', 'b', 'r'], marker='<', s=150)
````

````python
N = 1000
x = np.random.rand(N)
y = np.random.randn(N)
plt.scatter(x, y, color='red', alpha=0.3, s=5)
plt.xlim([-.5, 1.5])
````

На жаль, не можна (за допомогою одного виклику функції `scater`) зобразити точки з різними маркерами.

````python
z = np.array([0, 1]*(N//2))
colors = np.array(['red', 'blue'])[z]
markers = np.array(['v', '>'])[z]
plt.scatter(x, y, color=colors, alpha=0.55)
plt.xlim([-.5, 1.5])
````

## Лінії рівня

````python
X, Y = np.meshgrid([1, 2, 3, 4], [11, 22, 33, 44])
print(X)
print(Y)
````

Рассмотрим функцию
$$
z = (x^2 + y^2)^3 - 4x^2 y^2
$$

````python
x = np.linspace(-1, 1, 300)
X, Y = np.meshgrid(x, x)
````

````python
print(X.shape, Y.shape)
````

````python
x = np.linspace(-1, 1, 300)
X, Y = np.meshgrid(x, x)
Z = (X**2 + Y**2)**3 - 4*X**2.*Y**2
cv = [-0.1, -0.05, -0.01, -0.001, -0.00005, 0, 0.1]
plt.contour(X, Y, Z, cv)
plt.set_cmap('hot')
plt.colorbar()
````

````python
plt.contourf(X, Y, Z, cv)
````

````python
X, Y = np.meshgrid(np.arange(-3, 3, .1), np.arange(-3, 3, .1))
Z = np.sin(X) * np.sin(Y)
cs = plt.contour(X, Y, Z)
plt.clabel(cs, fontsize=10)
plt.set_cmap('hsv')
plt.colorbar()
````
