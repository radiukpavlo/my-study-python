```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score

def simulate_pr_data(n_samples: int, target_ap: float):
    """
    Simulates binary labels and prediction scores that result in a
    Precision-Recall curve with a desired shape and Average Precision (AP).

    Args:
        n_samples: The total number of samples to generate.
        target_ap: The target Average Precision (approx. 0.73 for this case).

    Returns:
        A tuple of (true_labels, predicted_scores).
    """
    # A higher ratio of negative samples makes the classification task harder.
    n_positives = int(n_samples * 0.35)
    n_negatives = n_samples - n_positives

    y_true = np.array([1] * n_positives + [0] * n_negatives)

    # Tune the means of the score distributions to get closer to the target AP.
    # A smaller gap between means results in a lower AP.
    mean_gap = 0.35 + (target_ap - 0.73)
    
    pos_scores = np.random.normal(loc=0.5 + mean_gap, scale=0.3, size=n_positives)
    neg_scores = np.random.normal(loc=0.5 - mean_gap, scale=0.25, size=n_negatives)
    
    y_scores = np.concatenate([pos_scores, neg_scores])
    y_scores = np.clip(y_scores, 0, 1) # Ensure scores are in [0, 1]

    # Shuffle the labels and scores together
    p = np.random.permutation(len(y_true))
    return y_true[p], y_scores[p]

def create_pr_curve_plot():
    """
    Generates and saves a Precision-Recall curve plot that precisely
    recreates the provided image.
    """
    # --- 1. Setup Plot Style and Font ---
    plt.style.use('ggplot')
    font_properties = {
        'family': 'sans-serif',
        'weight': 'bold',
        'size': 18
    }
    plt.rc('font', **font_properties)
    plt.rc('axes', titleweight='bold', labelweight='bold', titlesize=22, labelsize=20)
    plt.rc('xtick', labelsize=16)
    plt.rc('ytick', labelsize=16)
    plt.rc('legend', fontsize=16)
    plt.rc('axes', edgecolor='black', linewidth=1.5) # Set black border

    # --- 2. Generate Data and Metrics ---
    np.random.seed(42)
    y_true, y_scores = simulate_pr_data(n_samples=5000, target_ap=0.73)
    
    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    ap_score = average_precision_score(y_true, y_scores)

    # --- 3. Create the Plot ---
    fig, ax = plt.subplots(figsize=(8, 6))

    # Plot the thin "Building" line
    ax.plot(
        recall,
        precision,
        label=f'Building {ap_score:.3f}',
        linewidth=1.5 # Default thin line
    )

    # Plot the thick "all classes" line on top
    ax.plot(
        recall,
        precision,
        label=f'all classes {ap_score:.3f} mAP@0.5',
        linewidth=4,  # Thick line
        color='blue'
    )
    
    # --- 4. Finalize and Save Plot ---
    ax.set_title('Precision-Recall Curve')
    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')
    
    # Set axis limits and remove top/right grid lines for a clean look
    ax.set_xlim(0, 1.0)
    ax.set_ylim(0, 1.0)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # Place legend outside the plot area
    ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0.)

    # Add a thin black line at the bottom to match the original image
    ax.axhline(y=0, color='black', linewidth=1.5)

    output_filename = '64_fig_A_3.pdf'
    plt.savefig(output_filename, format='pdf', bbox_inches='tight')
    print(f"Precision-Recall curve successfully saved as '{output_filename}'")

if __name__ == '__main__':
    create_pr_curve_plot()
```
