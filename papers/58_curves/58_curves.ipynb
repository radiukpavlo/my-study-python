```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Remake of a pie chart distribution of C_final for 10 evaluated defects, showing
near alignment with expert ratings. This script adheres to PEP 8 guidelines
and saves the figure as a PDF.
"""

import matplotlib.pyplot as plt

def main():
    """
    Creates and saves a pie chart (58_fig_4_pie.pdf) illustrating
    the distribution of C_final among three defect categories.
    """
    # Defect labels (in Ukrainian) and their respective percentages
    labels = ["Crack", "Corrosion", "Cut"]
    sizes = [40, 30, 30]

    # Define a simple, readable color palette
    colors = ["#8e44ad", "#9b59b6", "#c39bd3"]

    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(5, 5), dpi=150)

    # Plot the pie chart with a starting angle and some styling
    wedges, text_labels, autopcts = ax.pie(
        sizes,
        labels=labels,
        colors=colors,
        autopct="%1.1f%%",
        startangle=140,
        wedgeprops={"edgecolor": "white", "linewidth": 1}
    )

    # Make the pie chart circular
    ax.axis("equal")

    # Tight layout and save as PDF
    plt.tight_layout()
    plt.savefig("58_fig_4_pie.pdf", format="pdf")
    plt.show()

if __name__ == "__main__":
    main()
```
