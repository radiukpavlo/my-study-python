```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the five arrays of 2 by 11.
anger = np.random.randint(0, 100, (2, 11))
fear = np.random.randint(0, 100, (2, 11))
joy = np.random.randint(0, 100, (2, 11))
neutral = np.random.randint(0, 100, (2, 11))
sadness = np.random.randint(0, 100, (2, 11))

# Output the five arrays to CSV files.
np.savetxt("anger.csv", anger, delimiter=",")
np.savetxt("fear.csv", fear, delimiter=",")
np.savetxt("joy.csv", joy, delimiter=",")
np.savetxt("neutral.csv", neutral, delimiter=",")
np.savetxt("sadness.csv", sadness, delimiter=",")

# Visualize the five arrays in Python.
fig, axes = plt.subplots(5, 2, figsize=(10, 10))
for i in range(5):
    axes[i, 0].scatter(anger[0], anger[1], label="Anger")
    axes[i, 1].scatter(fear[0], fear[1], label="Fear")
    axes[i, 0].scatter(joy[0], joy[1], label="Joy")
    axes[i, 1].scatter(neutral[0], neutral[1], label="Neutral")
    axes[i, 0].scatter(sadness[0], sadness[1], label="Sadness")
    axes[i, 0].set_title(f"Emotion {i}")
    axes[i, 1].set_title(f"Emotion {i}")
    axes[i, 0].legend()
    axes[i, 1].legend()
plt.show()
```
