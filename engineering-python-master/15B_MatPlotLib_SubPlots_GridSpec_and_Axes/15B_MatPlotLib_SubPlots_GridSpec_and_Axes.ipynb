# MatPlotLib SubPlots, GridSpec, and Axes

### Subplots

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 15)
C = np.cos(X)
S = np.sin(X)
```

```python
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.subplot(2, 1, 1)
plt.plot(X, C)
plt.subplot(2, 1, 2)
plt.plot(X, S)
plt.show()
```

```python
fig = plt.figure(figsize=(5, 4), dpi=80)
plt.subplot(1, 2, 1)
plt.plot(X, C)
plt.grid(False)
plt.subplot(1, 2, 2)
plt.plot(X, S)
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(6, 4), dpi=80)
plt.subplot(2, 2, 1)
plt.plot(X, C,'b-*')
plt.subplot(2, 2, 2)
plt.plot(X, S,'rs')
plt.subplot(2, 2, 3)
plt.plot(X, -C,'g--v')
plt.subplot(2, 2, 4)
plt.plot(X, -S, 'm:o')
plt.show()
```

### GridSpec

- GridSpec can define a subplot across multiple cells.
<img src="images/gridspec.png" alt="GridSpec Illustration" style="width: 200px;"/>

```python
import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(10, 5), dpi=80)
gs = gridspec.GridSpec(3, 3)
ax1 = plt.subplot(gs[0, :])
plt.plot(X, C)
ax2 = plt.subplot(gs[1, :-1])
plt.plot(X, S)
ax3 = plt.subplot(gs[-1, 0])
plt.plot(X, -C)
ax4 = plt.subplot(gs[-1, -2])
plt.plot(X, -S)
ax5 = plt.subplot(gs[1:, -1])
plt.plot(X, (C+S)/2)
plt.show()
```

### Axes

- Axes allow placement of plots at any location in the figure.

```python
fig = plt.figure(figsize=(5, 4), dpi=80)
ax1 = plt.axes([0.1, 0.1, 0.8, 0.8])  #[left, bottom, width, height])
plt.plot(X, S)
ax2 = plt.axes([0.2, 0.6, 0.3, 0.3])
plt.plot(X, C)
plt.show()
```
