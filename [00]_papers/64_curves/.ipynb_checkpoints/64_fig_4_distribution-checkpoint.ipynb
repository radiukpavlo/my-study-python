```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress warnings from seaborn that might arise with certain data
warnings.filterwarnings("ignore")

def create_distribution_plot():
    """
    Generates and saves a histogram and KDE plot that recreates the
    provided image of building size distributions.
    """
    # --- 1. Setup Plot Style and Font ---
    plt.style.use('ggplot')

    # Define font properties for bold and large text
    font_properties = {
        'family': 'sans-serif',
        'weight': 'bold',
        'size': 16  # Base font size
    }
    plt.rc('font', **font_properties)
    plt.rc('axes', titleweight='bold', labelweight='bold', titlesize=20, labelsize=18)
    plt.rc('xtick', labelsize=14)
    plt.rc('ytick', labelsize=14)
    plt.rc('legend', fontsize=16)

    # --- 2. Simulate Data to Match Original Plot ---
    # The data resembles a log-normal distribution. We'll tune the parameters
    # (mu and sigma of the underlying normal distribution) to match the visual.
    np.random.seed(42)
    n_samples = 2000

    # Satellite data: wider distribution, peak around 70
    satellite_mu, satellite_sigma = 4.1, 0.6
    satellite_sizes = np.random.lognormal(
        mean=satellite_mu, sigma=satellite_sigma, size=n_samples
    )

    # Drone data: narrower distribution, peak around 50
    drone_mu, drone_sigma = 3.9, 0.4
    drone_sizes = np.random.lognormal(
        mean=drone_mu, sigma=drone_sigma, size=n_samples
    )

    # --- 3. Create the Plot ---
    fig, ax = plt.subplots(figsize=(10, 7))

    # Define colors to match the original plot
    satellite_color = '#4c72b0'  # A muted blue
    drone_color = '#dd8452'      # A muted orange

    # Plot Satellite distribution
    sns.histplot(
        satellite_sizes,
        kde=True,
        stat="density",
        bins=25,
        color=satellite_color,
        label='Satellite',
        alpha=0.5,
        ax=ax,
        line_kws={'linewidth': 2.5}
    )

    # Plot Drone distribution
    sns.histplot(
        drone_sizes,
        kde=True,
        stat="density",
        bins=25,
        color=drone_color,
        label='Drone',
        alpha=0.5,
        ax=ax,
        line_kws={'linewidth': 2.5}
    )

    # --- 4. Finalize and Save Plot ---
    ax.set_title('Histogram + KDE of building sizes')
    ax.set_xlabel('Average bounding-box size (px)')
    ax.set_ylabel('Probability density')
    ax.legend(edgecolor='black', framealpha=1)
    
    # Set axis limits to match the original image for a clean look
    ax.set_xlim(0, 550)
    ax.set_ylim(0, 0.021)
    ax.grid(True, which='both', linestyle='--', linewidth=0.8)

    output_filename = '64_fig_4_distribution.pdf'
    plt.savefig(output_filename, format='pdf', bbox_inches='tight')
    print(f"Plot successfully saved as '{output_filename}'")

if __name__ == '__main__':
    create_distribution_plot()
```
