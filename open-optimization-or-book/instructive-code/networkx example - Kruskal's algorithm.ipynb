# Minimum Spanning trees

Lets create a minimum spanning tree using Kruskal's algorithm

```python
import networkx as nx

# one possible way of creating a graph
distance = [[0,4,3,6,1],
            [0,0,8,5,4],
            [0,0,0,2,12],
            [0,0,0,0,11],
            [0,0,0,0,0]]

nodes = ['a','b','c','d','e']

G = nx.Graph()

for v in nodes:
    G.add_node(v)

for i in range(len(nodes)):
    for j in range(i,len(nodes)):
        G.add_edge(nodes[i],nodes[j], weight = distance[i][j])

    
```

# Plotting the graph

```python
# Plot the graph
edge_labels = nx.get_edge_attributes(G, 'weight')
pos = nx.shell_layout(G) 

nx.draw(G,pos)
nx.draw_networkx_edge_labels(G,pos, edge_labels, font_size=14, rotate=False, font_color = 'b', font_weight = 'bold')
nodes = nx.draw_networkx_nodes(G, pos, node_size = 500, node_color = 'lightblue')
nodes.set_edgecolor('black')
nx.draw_networkx_labels(G, pos,  font_size = 16);
```

# Use networkx to solve for the minimum spanning tree

```python
? nx.minimum_spanning_tree
```

```python
# compute the minimum spanning tree
H = nx.minimum_spanning_tree(G)

# Plot the graph
edge_labels = nx.get_edge_attributes(H, 'weight')
pos = nx.shell_layout(G)

nx.draw(H,pos)
nx.draw_networkx_edge_labels(H,pos, edge_labels, font_size=14, rotate=False, font_color = 'b', font_weight = 'bold')
nodes = nx.draw_networkx_nodes(H, pos, node_size = 500, node_color = 'lightblue')
nodes.set_edgecolor('black')
nx.draw_networkx_labels(H, pos,  font_size = 16);
```

```python
edgelist = [('e','d')]

edge_labels = nx.get_edge_attributes(G, 'weight')
pos = nx.shell_layout(G)
nx.draw(G,pos)
nx.draw_networkx_edge_labels(G,pos, edge_labels, font_size=14, rotate=False, font_color = 'b', font_weight = 'bold')
nx.draw_networkx_edges(G,pos,width=8,alpha=0.5,edge_color="tab:red", edgelist=edgelist)
nodes = nx.draw_networkx_nodes(G, pos, node_size = 500, node_color = 'lightblue',node_shape='o')
nodes.set_edgecolor('black')
nx.draw_networkx_labels(G, pos,  font_size = 16);
```

## Make  this into a function

```python
edgelist = [('e','d')]

def highlight(edgelist):
    edge_labels = nx.get_edge_attributes(G, 'weight')
    pos = nx.shell_layout(G)
    nx.draw(G,pos)
    nx.draw_networkx_edge_labels(G,pos, edge_labels, font_size=14, rotate=False, font_color = 'b', font_weight = 'bold')
    nx.draw_networkx_edges(G,pos,width=8,alpha=0.5,edge_color="tab:red", edgelist=edgelist)
    nodes = nx.draw_networkx_nodes(G, pos, node_size = 500, node_color = 'lightblue',node_shape='o')
    nodes.set_edgecolor('black')
    nx.draw_networkx_labels(G, pos,  font_size = 16);

highlight(edgelist)
```

# Kruskal's Algorithm 
We now show how to do the steps of Kruskal's algorithm

Kruskal Algorithm Steps:
1.  Find edge of minimum weight
 - edge (a,e)
 
2.  Add edge (c,d)
3. Add edge (a,c)
4. Add edge (a,b)
5. Stop since the subgraph is now connected

(when doing this, report and rejected edges due to cycles)

```python
highlight([('a','e'), ('c','d'), ('a','c'), ('a','b')])
```

```python
highlight([('a','e'), ('c','d'), ('a','c'), ('a','b')])
```

1. Pick edge with smallest weight

```python
highlight([('a','e')])
```

2. Pick next edge as the edge that has the next smallest weight but doesn't create a cycle

Note: This edge does not need to be connected to the prior edge(s) chosen

```python
highlight([('a','e'), ('d','c')])
```

```python
highlight([('a','e'), ('d','c'), ('a', 'c')])
```

```python
highlight([('a','e'), ('d','c'), ('a', 'c'), ('a', 'b')])
```

Now that all nodes are connected, we are done.  
Notice that if we add any more edges, we will create a cycle

The minimum spanning tree has weight 1 + 4 + 2 + 3 = 10
