```python
import numpy as np
from sklearn.metrics import roc_auc_score

# True labels
y_true = np.array([0, 0, 1, 1])

# Predicted probabilities of class 1
y_scores = np.array([0.1, 0.4, 0.35, 0.8])

auc = roc_auc_score(y_true, y_scores)
print("AUC: ", auc)
```

```python
import numpy as np
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

# Assume that y_true is your true binary labels and y_scores are predicted probabilities for class 1
fpr, tpr, thresholds = roc_curve(y_true, y_scores)

plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc_score(y_true, y_scores))
plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
url = 'https://raw.githubusercontent.com/radiukpavlo/my-operations-research/main/data/fer_results.csv'
df = pd.read_csv(url)

# List of unique emotions
emotions = df['Emotion'].unique()

# For each emotion
for emotion in emotions:
    # Filter data for the current emotion
    df_emotion = df[df['Emotion'] == emotion]
    
    # Melt the dataframe
    df_melted = df_emotion.melt(id_vars='Matrix', value_vars=['Precision', 'Recall', 'F1-Score', 'Accuracy'], var_name='Metric', value_name='Value')
    
    # Create a barplot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Metric', y='Value', hue='Matrix', data=df_melted, palette='viridis')
    
    # Improve the legend 
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
    # Set the title
    plt.title(f'Comparison for Emotion: {emotion}')
    
    plt.show()
```
