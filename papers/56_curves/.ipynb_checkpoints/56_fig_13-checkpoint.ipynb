```python
import matplotlib.pyplot as plt
#plt.style.use('ggplot')
import numpy as np

# Data from the figure
categories = [
    "Myocardium (ES)", "Right ventricle (ES)", "Left ventricle (ES)",
    "Myocardium (ED)", "Right ventricle (ED)", "Left ventricle (ED)"
]
proposed = [0.920, 0.915, 0.940, 0.896, 0.947, 0.974]
exp4 = [0.898, 0.905, 0.930, 0.866, 0.939, 0.956]
exp3 = [0.885, 0.873, 0.887, 0.855, 0.892, 0.919]
exp2 = [0.884, 0.891, 0.894, 0.875, 0.902, 0.920]
exp1 = [0.832, 0.871, 0.890, 0.812, 0.842, 0.911]

# Bar positions
x = np.arange(len(categories))
width = 0.15

# Create the plot
fig, ax = plt.subplots(figsize=(15, 8))

# Plot bars with "Pythonic" style (rounded edges)
bars1 = ax.bar(x - 2 * width, proposed, width, label="Proposed approach", color="#2ca02c", edgecolor="black", linewidth=0.8)
bars2 = ax.bar(x - width, exp4, width, label="Experiment 4", color="#ff7f0e", edgecolor="black", linewidth=0.8)
bars3 = ax.bar(x, exp3, width, label="Experiment 3", color="#7f7f7f", edgecolor="black", linewidth=0.8)
bars4 = ax.bar(x + width, exp2, width, label="Experiment 2", color="#d62728", edgecolor="black", linewidth=0.8)
bars5 = ax.bar(x + 2 * width, exp1, width, label="Experiment 1", color="#1f77b4", edgecolor="black", linewidth=0.8)

# Add gridlines for better readability
ax.grid(axis='y', linestyle="--", alpha=0.7)

# Customize axes
ax.set_ylabel("Dice Coefficient", fontsize=16, fontweight="bold")
ax.set_xlabel("Categories", fontsize=16, fontweight="bold")
#ax.set_title("Visual Comparison of Dice Coefficient Values (ES and ED)", fontsize=18, fontweight="bold")
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=0, fontsize=14)
ax.set_ylim([0.7, 1.0])

# Customize Y-axis numbers (larger and bold)
ax.tick_params(axis='y', labelsize=14)
# for label in ax.get_yticklabels():
#     label.set_fontweight("bold")

# Customize X-axis numbers (larger and bold)
ax.tick_params(axis='x', labelsize=14)
# for label in ax.get_xticklabels():
#     label.set_fontweight("bold")

# Add legend
ax.legend(fontsize=14, title_fontsize=12, loc="upper left")

# Adjust layout to avoid text clipping
plt.tight_layout()

# Save the plot in SVG format
plt.savefig("56_fig_11.svg", format="svg")

# Display the plot
plt.show()
```
