```python
import pandas as pd
import matplotlib.pyplot as plt
import io

# Step 1: Store the CSV data and create the file
# This part remains unchanged and is for creating a self-contained example.
csv_data = """epoch,train/box_loss,train/seg_loss,train/cls_loss,train/dfl_loss,val/box_loss,val/seg_loss,val/cls_loss,val/dfl_loss
1,1.488,2.62549,1.91156,1.26968,1.33839,2.35881,1.45513,1.17854
2,1.42969,2.43641,1.42877,1.20764,1.36584,2.38583,1.42483,1.21843
3,1.47346,2.49803,1.41715,1.22796,1.43081,2.6062,1.56048,1.25233
4,1.48494,2.50379,1.38942,1.24269,1.3798,2.40785,1.31721,1.20992
5,1.45378,2.45114,1.32737,1.22878,1.35136,2.31372,1.28302,1.21815
6,1.4348,2.41381,1.29162,1.22042,1.31187,2.2537,1.18204,1.1761
7,1.41644,2.37122,1.24956,1.20558,1.29615,2.26084,1.15637,1.16768
8,1.40222,2.3534,1.23006,1.20314,1.31507,2.19598,1.16489,1.18291
9,1.38481,2.32515,1.20464,1.19178,1.27549,2.17508,1.128,1.14741
10,1.37831,2.30976,1.19292,1.1873,1.25948,2.16397,1.12629,1.16485
11,1.36759,2.28809,1.17794,1.18241,1.27752,2.16778,1.12877,1.15297
12,1.36327,2.27255,1.16101,1.17837,1.24732,2.1334,1.07925,1.15104
13,1.35838,2.26633,1.15184,1.17457,1.23166,2.10399,1.0579,1.14039
14,1.3462,2.24912,1.14079,1.1711,1.23313,2.08749,1.04954,1.13115
15,1.34261,2.24324,1.13339,1.16968,1.21905,2.07507,1.05538,1.13097
16,1.34482,2.22847,1.12609,1.16556,1.22099,2.08513,1.04154,1.1282
17,1.33071,2.21665,1.1137,1.1599,1.20535,2.05362,1.03742,1.12171
18,1.32813,2.21521,1.10973,1.16093,1.20058,2.04403,1.01675,1.11547
19,1.32351,2.20536,1.1042,1.15523,1.19658,2.04123,1.01847,1.1131
20,1.31783,2.18954,1.09845,1.1551,1.19438,2.02922,1.00411,1.11696
21,1.31551,2.1818,1.09621,1.15205,1.18228,2.01169,0.99728,1.10914
22,1.31057,2.17553,1.08029,1.14881,1.18546,2.01595,0.99708,1.10888
23,1.30712,2.17721,1.07664,1.14636,1.18464,2.00663,0.98434,1.10806
24,1.30426,2.16501,1.07268,1.14616,1.18443,1.99931,0.98129,1.1032
25,1.30252,2.15842,1.06362,1.14484,1.17571,1.99334,0.97149,1.10164
26,1.29649,2.15055,1.06382,1.1441,1.17652,1.9973,0.98229,1.103
27,1.29225,2.14945,1.05488,1.14051,1.17377,1.98297,0.96126,1.10304
28,1.29775,2.15046,1.05651,1.14102,1.16295,1.97459,0.9668,1.09725
29,1.28789,2.13778,1.05433,1.13852,1.16824,1.98103,0.9613,1.09869
30,1.28704,2.12725,1.04573,1.13476,1.16002,1.96602,0.9521,1.09583
31,1.28853,2.13803,1.0433,1.1351,1.16075,1.97241,0.95718,1.0927
32,1.28669,2.13209,1.04032,1.13457,1.15598,1.95354,0.9469,1.09319
33,1.27545,2.11918,1.03681,1.13304,1.15773,1.9526,0.94204,1.09305
34,1.27564,2.11381,1.03194,1.12841,1.15552,1.95094,0.94144,1.09151
35,1.27316,2.1106,1.02778,1.12803,1.15009,1.94919,0.94149,1.08736
36,1.27139,2.10469,1.0259,1.12771,1.14742,1.9451,0.93459,1.08894
37,1.2691,2.09752,1.01807,1.12194,1.14872,1.94053,0.92959,1.08925
38,1.27261,2.09987,1.01868,1.12539,1.14912,1.94182,0.92938,1.08742
39,1.26509,2.09188,1.01345,1.12438,1.1443,1.93257,0.92666,1.08705
40,1.26038,2.08637,1.01103,1.12141,1.14619,1.93831,0.92944,1.0862
41,1.25963,2.08275,1.01344,1.12122,1.14308,1.93205,0.9276,1.08725
42,1.25766,2.08895,1.0058,1.11936,1.14039,1.93011,0.91956,1.08376
43,1.25524,2.08339,1.00217,1.11931,1.13948,1.92435,0.9209,1.0834
44,1.2587,2.07209,1.0014,1.11734,1.1379,1.92624,0.92196,1.081
45,1.25358,2.07466,0.99852,1.11607,1.13571,1.92541,0.92121,1.08047
46,1.24946,2.06675,0.99382,1.11509,1.13682,1.92443,0.91798,1.07974
47,1.24714,2.05581,0.99063,1.11371,1.13581,1.92288,0.91545,1.08073
48,1.24492,2.05488,0.98445,1.11223,1.13796,1.92447,0.91561,1.0823
49,1.24269,2.05285,0.98574,1.1124,1.13656,1.92314,0.91505,1.08118
50,1.24282,2.05672,0.98093,1.11027,1.1339,1.92237,0.91051,1.0808
51,1.24141,2.04832,0.98024,1.10918,1.13265,1.91983,0.90963,1.0794
52,1.24001,2.0519,0.97903,1.10962,1.13203,1.91732,0.90948,1.07889
53,1.23975,2.04546,0.97673,1.10725,1.1329,1.91653,0.91107,1.07892
54,1.23685,2.04325,0.97326,1.10745,1.13235,1.91336,0.9093,1.07782
55,1.2329,2.0344,0.96971,1.10386,1.13112,1.91251,0.90832,1.07723
56,1.23187,2.03221,0.96024,1.10335,1.13186,1.91397,0.90836,1.07748
57,1.22794,2.02584,0.9566,1.10225,1.13099,1.91563,0.90809,1.07691
58,1.22938,2.0309,0.95908,1.10303,1.12983,1.9136,0.90635,1.07758
59,1.22193,2.02102,0.95655,1.10174,1.12813,1.91381,0.90592,1.07694
60,1.22548,2.02165,0.95109,1.10166,1.12853,1.91198,0.90592,1.07728
61,1.22825,2.01443,0.95039,1.10009,1.12907,1.9122,0.90596,1.07742
62,1.21792,2.01093,0.9413,1.09731,1.12963,1.91227,0.90531,1.07712
63,1.21879,2.00922,0.94348,1.09718,1.12913,1.91233,0.90422,1.0772
64,1.21672,2.01156,0.94507,1.0959,1.1294,1.91095,0.90431,1.07716
65,1.2128,2.00541,0.93929,1.09546,1.12892,1.9104,0.90314,1.07694
66,1.2175,2.00639,0.93762,1.09651,1.12983,1.91047,0.90255,1.07745
67,1.21056,1.99667,0.93282,1.09173,1.12999,1.91051,0.90157,1.0772
68,1.20759,1.98904,0.93068,1.092,1.13011,1.91117,0.90163,1.07669
69,1.205,1.98608,0.9239,1.09078,1.13007,1.91135,0.90125,1.07645
70,1.2127,1.98964,0.92607,1.09261,1.12968,1.91167,0.90012,1.07615
71,1.19862,1.97772,0.92163,1.08991,1.12983,1.91167,0.90001,1.07631
72,1.19575,1.97381,0.92022,1.08831,1.13019,1.91142,0.90036,1.07657
73,1.1987,1.97546,0.9173,1.08641,1.13013,1.91184,0.90026,1.0764
74,1.1998,1.97452,0.91107,1.08767,1.13028,1.91229,0.89995,1.07623
75,1.19383,1.95873,0.90669,1.08288,1.13027,1.91217,0.90003,1.07614
76,1.19723,1.97057,0.90609,1.08522,1.1303,1.9121,0.9,1.07611
77,1.19153,1.95736,0.89907,1.08155,1.13035,1.91178,0.90011,1.07605
78,1.19044,1.95655,0.90575,1.0811,1.13047,1.91257,0.9003,1.07618
79,1.18497,1.95843,0.89823,1.08051,1.13051,1.91272,0.90014,1.07619
80,1.18627,1.94831,0.89578,1.07957,1.13102,1.91299,0.90017,1.07661
81,1.18589,1.94799,0.89365,1.07941,1.13122,1.91312,0.89991,1.07663
82,1.18042,1.94168,0.89108,1.07608,1.13172,1.91329,0.89971,1.07707
83,1.17825,1.93704,0.88562,1.07558,1.1316,1.91412,0.8999,1.07694
84,1.17873,1.93414,0.88368,1.07516,1.13166,1.91447,0.90009,1.07688
85,1.17531,1.92822,0.87848,1.07454,1.13191,1.91453,0.90009,1.07708
86,1.17812,1.93047,0.87981,1.07298,1.13219,1.91518,0.9001,1.07728
87,1.17134,1.92592,0.87248,1.07135,1.13263,1.91513,0.90021,1.07753
88,1.17234,1.92527,0.87175,1.07063,1.13291,1.91544,0.9004,1.07745
89,1.16685,1.91301,0.8662,1.07016,1.13292,1.91564,0.90065,1.07743
90,1.16608,1.90723,0.86377,1.06784,1.13328,1.91589,0.90068,1.07742
91,1.14693,1.89139,0.84238,1.07249,1.13362,1.9162,0.90094,1.07765
92,1.14261,1.88673,0.83124,1.06973,1.13388,1.9162,0.90089,1.07761
93,1.13625,1.86666,0.82355,1.06809,1.13398,1.9162,0.90107,1.07767
94,1.13291,1.86243,0.81529,1.06491,1.13418,1.91669,0.90104,1.07775
95,1.12934,1.8552,0.81447,1.06423,1.13431,1.91647,0.90112,1.07787
96,1.12505,1.85172,0.81026,1.06311,1.1345,1.91671,0.90128,1.07791
97,1.12014,1.84761,0.80432,1.0586,1.13477,1.91733,0.90138,1.07813
98,1.11868,1.83954,0.79985,1.05839,1.13504,1.91725,0.90158,1.07824
99,1.11606,1.83945,0.79827,1.05678,1.13537,1.91758,0.90165,1.07845
100,1.11301,1.83028,0.79239,1.0561,1.13586,1.91814,0.90196,1.07868
"""
file_name = 'finetunning_yolo_generated_results_for_losses.csv'
with open(file_name, 'w') as f:
    f.write(csv_data)

print(f"'{file_name}' created successfully.")

# Step 2: Load the data from the CSV file
df = pd.read_csv(file_name)

# Identify the columns for training and validation losses
train_loss_cols = [col for col in df.columns if col.startswith('train/')]
val_loss_cols = [col for col in df.columns if col.startswith('val/')]

# REFACTORED plotting function with enhanced text properties and independent axes
def create_loss_curves_figure_enhanced(data, columns, save_path):
    """
    Generates and saves a figure with enhanced text and independent y-axes
    for each subplot, arranged in a single row.
    
    Args:
        data (pd.DataFrame): The dataframe containing the training data.
        columns (list): A list of column names to plot.
        save_path (str): The path to save the generated PDF file.
    """
    num_plots = len(columns)
    
    # Define font properties for consistency
    TITLE_FONT = {'size': '18', 'weight': 'bold'}
    LABEL_FONT = {'size': '16', 'weight': 'bold'}
    TICK_FONT_SIZE = 14
    LEGEND_FONT_SIZE = 14
    FONT_WEIGHT = 'bold'
    
    # Create a figure with independent Y-axes by setting sharey=False (default)
    fig, axes = plt.subplots(1, num_plots, figsize=(20, 4))
    
    if num_plots == 1:
        axes = [axes]
        
    print(f"Generating enhanced plot for '{save_path}'...")

    for i, col_name in enumerate(columns):
        ax = axes[i]
        
        # Plot the raw and smoothed results
        ax.plot(data['epoch'], data[col_name], marker='o', linestyle='-', markersize=4, label='results')
        smooth_data = data[col_name].rolling(window=5, min_periods=1, center=True).mean()
        ax.plot(data['epoch'], smooth_data, linestyle='--', color='darkorange', label='smooth')
        
        # Set titles and labels for each subplot using the defined font properties
        ax.set_title(col_name, fontdict=TITLE_FONT)
        ax.set_xlabel('Epoch', fontdict=LABEL_FONT)
        ax.set_ylabel('Loss', fontdict=LABEL_FONT)
        
        # Style the legend
        legend = ax.legend(fontsize=LEGEND_FONT_SIZE)
        for text in legend.get_texts():
            text.set_fontweight(FONT_WEIGHT)
        
        # Style the tick labels to be larger and bold
        ax.tick_params(axis='both', which='major', labelsize=TICK_FONT_SIZE)
        for label in (ax.get_xticklabels() + ax.get_yticklabels()):
            label.set_fontweight(FONT_WEIGHT)
            
        ax.grid(True, linestyle='--', alpha=0.6)

    # Adjust layout to prevent plots and labels from overlapping
    plt.tight_layout()
    
    # Save the figure to a PDF file
    plt.savefig(save_path, format='pdf', bbox_inches='tight')
    print(f"Figure saved to '{save_path}'")
    plt.close(fig)

# --- Generate and Save Enhanced Training Loss Figure ---
create_loss_curves_figure_enhanced(
    data=df,
    columns=train_loss_cols,
    save_path='64_fig_A_1_train.pdf'
)

# --- Generate and Save Enhanced Validation Loss Figure ---
create_loss_curves_figure_enhanced(
    data=df,
    columns=val_loss_cols,
    save_path='64_fig_A_1_valid.pdf'
)
```
