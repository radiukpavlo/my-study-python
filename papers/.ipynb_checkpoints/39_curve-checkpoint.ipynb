```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import learning_curve
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Задамо графіки у форматі .svg, щоби вони мали кращу чіткість
%config InlineBackend.figure_format = 'svg' 

# Підвищимо розмір графіків за замовчуванням
# plt.style.use('ggplot')
# plt.rcParams['figure.figsize'] = 8, 5
# plt.rcParams['axes.grid'] = True
```

```python
my_data_acc = pd.read_excel('39_data.xlsx', sheet_name='Sheet2')

# Extract data from the specified cells
acc_epochs = my_data_acc.iloc[0:30, 0].values
acc_train = my_data_acc.iloc[0:30, 3].values
acc_valid = my_data_acc.iloc[0:30, 4].values

# Plot the data
plt.figure(figsize=(5,4), dpi=600)
plt.rcParams.update({'font.size': 10, 'font.weight': 'bold'})

plt.plot(acc_epochs, acc_train, color='blue', marker='x', markersize=4, label='Training')
plt.plot(acc_epochs, acc_valid, color='green', marker='D', markersize=4, linestyle='--', label='Validation')

plt.title('Accuracy learning curves', fontweight='bold', fontsize=14)
plt.yticks(fontweight='bold', fontsize=12)
#plt.xticks(np.arange(1, 32, step=4), fontweight='bold', fontsize=12)
plt.xlabel('Number of epochs', fontweight='bold', fontsize=14)
plt.ylabel('Accuracy', fontweight='bold', fontsize=14)

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend(loc='lower right', fontsize=14)

# Set the limits of y-axis
# plt.ylim(0.7,)
# plt.xlim(-1, 21)

# Save the plot
plt.savefig('39_fig_13a_curve.svg', format='svg', bbox_inches='tight')

# Display the plot
plt.show()
```

```python
my_data_loss = pd.read_excel('39_data.xlsx', sheet_name='Sheet2')

# Extract data from the specified cells
loss_epochs = my_data_loss.iloc[0:30, 0].values
loss_train = my_data_loss.iloc[0:30, 1].values
loss_valid = my_data_loss.iloc[0:30, 2].values

# Plot the data
plt.figure(figsize=(5,4), dpi=600)
plt.rcParams.update({'font.size': 10, 'font.weight': 'bold'})

plt.plot(loss_epochs, loss_train, color='blue', marker='x', markersize=4, label='Training')
plt.plot(loss_epochs, loss_valid, color='green', marker='D', markersize=4, linestyle='--', label='Validation')

plt.title('Loss learning curves', fontweight='bold', fontsize=14)
plt.yticks(fontweight='bold', fontsize=12)
# plt.xticks(np.arange(0, 34, step=4), fontweight='bold', fontsize=12)
plt.xlabel('Number of epochs', fontweight='bold', fontsize=14)
plt.ylabel('Loss value', fontweight='bold', fontsize=14)

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend(loc='upper right', fontsize=14)

# Set the limits of y-axis
# plt.ylim(0.7,)
# plt.xlim(-1, 21)

# Save the plot
plt.savefig('39_fig_13b_curve.svg', format='svg', bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# Load the Excel file
my_data = pd.read_excel('39_data.xlsx', sheet_name='Sheet1')

# Extract data from the specified cells
my_threshold = my_data.iloc[0:10, 0].values  # Assuming 'A' column is at index 0
metric_precision = my_data.iloc[0:10, 1].values  # Assuming 'B' column is at index 1
metric_recall = my_data.iloc[0:10, 2].values  # Assuming 'C' column is at index 2
metric_map = my_data.iloc[0:10, 3].values  # Assuming 'D' column is at index 3

# Plot the data
plt.figure(figsize=(8,4), dpi=600)
plt.rcParams.update({'font.size': 10, 'font.weight': 'bold'})

plt.plot(my_threshold, metric_precision, label='Precision', color='blue', marker='x', markersize=5)
plt.plot(my_threshold, metric_recall, label='Recall', color='green', marker='o', markersize=5)
plt.plot(my_threshold, metric_map, label='mAP', color='red', marker='D', markersize=5)

# Set the labels and title
# plt.xlabel('Threshold')
# plt.ylabel('Metric Value')
# plt.title('Visualization of Precision, Recall, and MAP')
# plt.legend()
# plt.grid(True)
# plt.tight_layout()

# plt.title('Accuracy learning curves', fontweight='bold', fontsize=14)
plt.yticks(fontweight='bold', fontsize=12)
plt.xlabel('Threshold of confidence', fontweight='bold', fontsize=12)
plt.ylabel('Metrics value', fontweight='bold', fontsize=12)

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend(loc='lower center', fontsize=12)

# Save the plot
plt.savefig('39_fig_15_threshold.svg', format='svg', bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# Data
labels = ['Sunny (the Sun is \n at its zenith)', 'Cloudy', 'Sunny (high shade)']
bar_precision = [0.92, 0.86, 0.78]
bar_recall = [0.89, 0.82, 0.74]
bar_f1score = [0.9, 0.84, 0.72]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(8,4))
plt.rcParams.update({'font.size': 10, 'font.weight': 'bold'})

# Soft colors for the bars
colors = ['#D6EAF8', '#85C1E9', '#2E86C1']

# Create bars
rects1 = ax.bar(x - width, bar_precision, width, label='Precision', color=colors[0])
rects2 = ax.bar(x, bar_recall, width, label='Recall', color=colors[1])
rects3 = ax.bar(x + width, bar_f1score, width, label='F1-Score', color=colors[2])

# Set y-axis limits
ax.set_ylim(0.70, 0.95)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Values', fontweight='bold', fontsize=12)
# ax.set_title('Evaluation of recognized fruits by a group of UAVs', fontweight='bold', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontweight='bold', fontsize=12)
ax.legend(loc='upper center', ncol=3, fontsize=12)

# Add grid lines
ax.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.6)
ax.set_axisbelow(True)  # Ensure grid is behind the bars

# Autolabel function to display the label on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontweight='bold')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)


fig.tight_layout()

# Save the plot
plt.savefig('39_fig_17_eval.svg', format='svg', bbox_inches='tight')

# Display the plot
plt.show()
```
