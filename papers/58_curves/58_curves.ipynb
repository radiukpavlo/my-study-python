```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Remake of a pie chart distribution of C_final for 10 evaluated defects,
showing near alignment with expert ratings. This script adheres to PEP 8
guidelines, maximizes chart space, and saves the figure as a PDF.
"""

import matplotlib.pyplot as plt

def main():
    """
    Creates and saves a pie chart (58_fig_4_pie.pdf) illustrating
    the distribution of C_final among three defect categories.
    """
    # Defect labels and respective percentages
    labels = ["Cracks", "Corrosion", "Overheating"]
    sizes = [40, 30, 30]

    # Define a simple, readable color palette
    colors = ["#8e44ad", "#9b59b6", "#c39bd3"]

    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(5, 5), dpi=150)

    # Plot the pie chart with bold, larger text and minimal whitespace
    wedges, text_labels, autopcts = ax.pie(
        sizes,
        labels=labels,
        colors=colors,
        autopct="%1.1f%%",
        startangle=140,
        wedgeprops={"edgecolor": "white", "linewidth": 1},
        textprops={"fontsize": 18, "fontweight": "bold"}
    )

    # Make the pie chart circular
    ax.axis("equal")

    # Remove extra margins
    fig.subplots_adjust(left=0, right=1, top=1, bottom=0)

    # Save the figure as PDF and SVG with tight bounding box
    plt.savefig("58_fig_4_pie.pdf", format="pdf", bbox_inches="tight", pad_inches=0.0)
    plt.savefig("58_fig_4_pie.svg", format="svg", bbox_inches="tight", pad_inches=0.0)
    plt.show()

if __name__ == "__main__":
    main()
```
