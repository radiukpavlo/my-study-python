```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add a vertical line
ax.axvline(x=5, color='r', linestyle='--')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Download the data frame
url = 'https://raw.githubusercontent.com/radiukpavlo/my-operations-research/main/data/df_results.csv'
df_results = pd.read_csv(url)

# Define the emotions and metrics
emotions = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
metrics = ['Precision', 'Recall', 'F1-Score', 'Accuracy']

# Plot histograms for each emotion and metric
for emotion in emotions:
    for metric in metrics:
        plt.figure(figsize=(8, 6))
        sns.barplot(x='Matrix', y=metric, data=df_results[df_results['Emotion'] == emotion], palette='Set2')
        plt.xlabel('Matrix')
        plt.ylabel(metric)
        plt.title(f'{emotion.capitalize()} - {metric}')
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.savefig(f'{emotion}_{metric}.svg', format='svg', bbox_inches='tight')
        plt.close()
```
