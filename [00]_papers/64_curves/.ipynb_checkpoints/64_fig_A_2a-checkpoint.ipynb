```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def create_confusion_matrix_plot():
    """
    Generates and saves a publication-quality confusion matrix plot that
    precisely recreates the provided image.
    """
    # --- 1. Setup Plot Style and Font ---
    plt.style.use('ggplot')
    font_properties = {
        'family': 'sans-serif',
        'weight': 'bold',
        'size': 20  # Base font size for high readability
    }
    plt.rc('font', **font_properties)
    plt.rc('axes', titleweight='bold', labelweight='bold', titlesize=22, labelsize=20)
    plt.rc('xtick', labelsize=18)
    plt.rc('ytick', labelsize=18)

    # --- 2. Define the Confusion Matrix Data ---
    # Data is digitized directly from the input image.
    # The structure is [[TP, FP], [FN, TN]]
    conf_matrix_data = np.array([
        [18794, 8462],
        [5628, 0]  # TN is 0, which will appear white with the 'Blues' colormap
    ])
    
    class_labels = ['Building', 'Background']

    # --- 3. Create the Heatmap Plot ---
    fig, ax = plt.subplots(figsize=(8, 7))

    # Use seaborn's heatmap for the base plot
    sns.heatmap(
        conf_matrix_data,
        annot=True,
        fmt='d',  # Format as integers
        cmap='Blues',
        linewidths=0.5,
        linecolor='gray',
        cbar=True,
        xticklabels=class_labels,
        yticklabels=class_labels,
        ax=ax,
        annot_kws={"size": 18} # Increase size of the numbers in cells
    )

    # --- 4. Customize Annotations for a Perfect Match ---
    # Seaborn adds text objects to the axes. We can access and modify them.
    # The text color should contrast with the cell background color.
    
    # Set a threshold for when to switch text color from black to white
    color_threshold = conf_matrix_data.max() / 2.0
    
    # The text objects are stored in ax.texts. We iterate through them.
    # The order corresponds to the flattened data array.
    for i, text_obj in enumerate(ax.texts):
        # Get the corresponding data value
        data_value = conf_matrix_data.flatten()[i]
        
        # Set text color to white for dark cells, black for light cells
        if data_value > color_threshold:
            text_obj.set_color('white')
        else:
            text_obj.set_color('black')
            
        # Explicitly hide the text for the cell with value 0 (bottom-right)
        if data_value == 0:
            text_obj.set_text("")

    # --- 5. Finalize and Save Plot ---
    ax.set_title('Confusion Matrix')
    ax.set_xlabel('True')
    ax.set_ylabel('Predicted')
    
    # Ensure y-axis labels are not rotated
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

    # Add a thin black line at the bottom to match the original image
    ax.axhline(y=2, color='black', linewidth=1.5)

    output_filename = '64_fig_A_2a.pdf'
    plt.savefig(output_filename, format='pdf', bbox_inches='tight')
    print(f"Confusion matrix plot successfully saved as '{output_filename}'")

if __name__ == '__main__':
    create_confusion_matrix_plot()
```
